# Comparisons

1. Vanilla:
- estimate the expected return by sampling single time steps (or transitions) and averaging the returns obtained from these steps.
- +: unbiased
- -: but has high variance

2. N-steps:
-  estimate the expected return by bootstrapping from the value of the state N steps ahead.
- +: reduces variance compared to Vanilla MC
- -: hard to choose the correct N, might change on different tasks

3. GAE
- combines the advantages of both MC and N-Steps by using a parameter λ to control the trade-off between them.
- +: reduces variance while maintaining a low bias
- -: hard to choose the correct $λ$ parameter, which can affect performance and thus require tuning

4. MC
-  estimate the expected return by averaging the returns obtained by running multiple complete episodes
- +: unbiased and consistent
- -: high variance
